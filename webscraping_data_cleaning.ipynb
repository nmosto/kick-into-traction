```python
# Load required libraries
import pandas as pd
import joblib
import json
import time
```

```python
# Initalize an empty DataFrame with labels of data to be collected
df = pd.DataFrame(
    columns=['name', 'id', 'blurb', 'category', 'url', 'currency', 'pledged', 'goal', 'state',
             'location','creator_id', 'creator_name' 'created_at', 'launched_at', 'deadline', 
             'backers', 'staff_pick']
)
```

```python
# Select JSON kickstarter data
filename = 'Kickstarter_2020-08-13T03_20_17_470Z.json'
```

```python
# Record start time
start = time.time()

# Open JSON data file
with open(filename, encoding='utf8') as file:
    for index, line in enumerate(file):
        # Read each line and convert the JSON String into a Python dictionary
        json_obj = json.loads(line)
        
        # Catch any potential typos or missing keys that can raise a KeyError
        try:
            df.loc[index, 'name'] = json_obj['data']['name']
        except KeyError:
            continue 
        
        try:
            df.loc[index, 'id'] = json_obj['data']['profile']['id']
        except KeyError:
            continue 
            
        try:
            df.loc[index, 'blurb'] = json_obj['data']['blurb']
        except KeyError:
            continue 
        
        try:
            df.loc[index, 'category'] = json_obj['data']['category']['name']
        except KeyError:
            continue
        
        try:
            df.loc[index, 'url'] = \
                json_obj['data']['urls']['web']['project']
        except KeyError:
            continue
        
        try:
            df.loc[index, 'currency'] = json_obj['data']['currency']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'pledged'] = json_obj['data']['pledged']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'goal'] = json_obj['data']['goal']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'state'] = json_obj['data']['state']
        except KeyError:
            continue
        
        try:
            df.loc[index, 'location'] = \
                json_obj['data']['location']['displayable_name']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'creator_id'] = json_obj['data']['creator']['id']
        except KeyError:
            continue 
        
        try:
            df.loc[index, 'creator_name'] = json_obj['data']['creator']['name']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'created_at'] = json_obj['data']['created_at']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'launched_at'] = json_obj['data']['launched_at']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'deadline'] = json_obj['data']['deadline']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'backers'] = json_obj['data']['backers_count']
        except KeyError:
            continue
            
        try:
            df.loc[index, 'staff_pick'] = json_obj['data']['staff_pick']
        except KeyError:
            continue
            
# Report elapsed time in seconds
time.time() - start
```
